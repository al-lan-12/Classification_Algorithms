# Predicting if the person will have diabetes or not based on input Using SVM

```python
import pandas as pd
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

# Importing Dataset

```python
df=pd.read_csv(r"C:\Users\joe\Downloads\Diabetes.csv")
df.head()
```

# Feature Selection

```python
x=df[['Pregnancies','Glucose','BloodPressure','Insulin','BMI','DiabetesPedigreeFunction','Age']]
y=df['Outcome']
print(x)
print(y)
```

# Splitting Testing and Training data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
x_train
```

```python
x_test
```

# Creating the Support Vector Machine Model

```python
clf=svm.SVC(kernel="linear")
clf=clf.fit(x_train,y_train)
```

# Predicting on test set

```python
y_pred=clf.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

# Model Evaluation

```python
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```
